```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2
import os
import numpy as np
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

batch_size = 32
```

```python
training_set = train_datagen.flow_from_directory('train/', target_size=(256, 256), batch_size=batch_size, class_mode='categorical', subset='training')

validation_set = train_datagen.flow_from_directory('train/', target_size=(256, 256), batch_size=batch_size, class_mode='categorical', subset='validation')
```

```python
training_set.class_indices
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16,(3,3),1,activation='relu',input_shape=(256,256,3)),
    tf.keras.layers.MaxPool2D(),
    tf.keras.layers.Conv2D(32,(3,3),1,activation='relu'),
    tf.keras.layers.MaxPool2D(),
    tf.keras.layers.Conv2D(16,(3,3),1,activation='relu'),
    tf.keras.layers.MaxPool2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256,activation='relu'),
    tf.keras.layers.Dense(5,activation='softmax'),
])
```

```python
model.compile( loss="categorical_crossentropy",optimizer="adam",metrics=['acc'])
```

```python
history = model.fit(training_set, epochs=10, validation_data=validation_set)
```
